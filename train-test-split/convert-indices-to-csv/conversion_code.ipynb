```python
import pandas as pd
```

```python
with open("test.txt",'r') as test:
    lines = test.readlines()
    line=lines[0]
    line = lines[0]
    arr = line.split(', ')
    arr[0] = arr[0][1:]
    arr[-1] = arr[-1][:-1]
    testids = [int(s) for s in arr]
```

```python
with open("train.txt",'r') as train:
    lines = train.readlines()
    line=lines[0]
    line = lines[0]
    arr = line.split(', ')
    arr[0] = arr[0][1:]
    arr[-1] = arr[-1][:-1]
    trainids = [int(s) for s in arr]
```

```python
with open("valn.txt",'r') as valn:
    lines = valn.readlines()
    line=lines[0]
    line = lines[0]
    arr = line.split(', ')
    arr[0] = arr[0][1:]
    arr[-1] = arr[-1][:-1]
    valids = [int(s) for s in arr]
```

```python
with open("final_bad_ids.txt",'r') as bad:
    lines = bad.readlines()
    line=lines[0]
    line = lines[0]
    arr = line.split(', ')
    arr[0] = arr[0][1:]
    arr[-1] = arr[-1][:-1]
    badids = [int(s) for s in arr]
```

```python
df = pd.read_excel("behaviour_simulation_train.xlsx")
```

```python
test_df = df[df['id'].isin(testids)]
train_df = df[df['id'].isin(trainids)]
val_df = df[df['id'].isin(valids)]
bad_df = df[df['id'].isin(badids)]
```
